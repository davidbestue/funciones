```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
d1 = np.random.randint(1,7)
d2 = np.random.randint(1,7)
d3 = np.random.randint(1,7)
dt = np.random.choice([1,3,4,5,5,6])
```

```python
## Results normal dice
```

```python
t_n=[]
for i in range(100000):
    d1 = np.random.randint(1,7)
    d2 = np.random.randint(1,7)
    d3 = np.random.randint(1,7)
    t_n.append(d1+d2+d3)

  
```

```python
sns.distplot(t_n, fit=norm, kde=True, norm_hist=True)
```

```python
np.array(t_n).mean()
```

```python
sum(np.array(t_n)==10)/100000 + sum(np.array(t_n)==11)/100000
```

```python
sum(np.array(t_n)<10.5)/100000
```

```python
sum(np.array(t_n)>10.5)/100000
```

```python
## Results with fake dice
```

```python
t_f=[]
for i in range(100000):
    d1 = np.random.randint(1,7)
    d2 = np.random.randint(1,7)
    dt = np.random.choice([1,3,4,5,5,6])
    t_f.append(d1+d2+dt)
```

```python
sns.distplot(t_f, fit=norm, kde=True, norm_hist=True)
```

```python
np.array(t_f).mean()
```

```python
sum(np.array(t_f)>11)/100000
```

```python
sum(np.array(t_f)<11)/100000
```

```python
round(sum(np.array(t_f)==11)/100000 + sum(np.array(t_f)==12)/100000, 2)
```

```python
round(sum(np.array(t_f)==11)/100000 + sum(np.array(t_f)==10)/100000, 2)
```

```python
sns.distplot(t_n, fit=norm, kde=False, norm_hist=True, color='olive', fit_kws={"color":'olive', 'linewidth':3}, label='normal')
sns.distplot(t_f, fit=norm, kde=False, norm_hist=True, color='navy', fit_kws={"color":'navy', 'linewidth':3}, label='gaff')
plt.legend()
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().get_xaxis().tick_bottom()
plt.gca().get_yaxis().tick_left()
plt.tick_params(direction='in')
```
