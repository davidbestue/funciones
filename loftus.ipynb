```python
values = [31.8, 34, 38.1, 39.3, 40.8]
values_kmh = np.array(values)* 1.60934
labels=['contactar', 'golpear', 'chocar', 'colisionar', 'estampar'] 
```

```python
# Estilo general
sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 5))
plt.grid(axis='x', visible=False)  # quitar líneas verticales
plt.grid(axis='y', linestyle='--', alpha=0.7)  # mantener horizontales

# Color azul oscuro
color = 'navy'  # azul oscuro

# Crear gráfico
bars = plt.bar(labels, values_kmh, color=color, edgecolor="black")
plt.xticks(fontsize=12, fontweight='bold')


# Etiquetas y título
plt.ylabel("Velocidad (km/h)", fontsize=12, fontweight='bold')
#plt.title("Velocidad Estimada para Cada Verbo", fontsize=14, fontweight='bold')

# Mostrar valores encima de las barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.8, f"{yval:.1f}", 
             ha='center', va='bottom', fontsize=10, fontweight='bold')

# Quitar bordes innecesarios
sns.despine()

plt.tight_layout()
plt.savefig("velocidades_verbos.png", dpi=300, bbox_inches='tight')
```
