```python
from linares_plot import *
```

```python
data = sns.load_dataset("titanic")
data.head()
```

```python
df=data[~data['age'].isnull()]
data_survived = df.loc[df['survived']==1]
```

```python
linares_plot(x='class', y='age', df=data_survived, 
             palette=['darkblue', 'darkred', 'darkgreen'],  
             order=['First', 'Second', 'Third'], 
             point_size=7, sep_axis=False,
             fill_box=True, alpha_box=0.4)
```

```python
linares_plot(x='class', y='age', df=data_survived.loc[data_survived['sex']=='male'], 
             palette=['darkblue', 'darkred', 'darkgreen'],  
             order=['First', 'Second', 'Third'], 
             point_size=7, sep_axis=False,
             fill_box=True, alpha_box=0.4)
plt.title('Men')
```
